```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Sample data
df = pd.DataFrame({
    'highway_mpg': [22, 25, 29, 32, 38],
    'price': [18000, 19000, 20000, 21000, 22000]
})

# Regression plot
sns.regplot(x='highway_mpg', y='price', data=df)
plt.title('Price vs Highway MPG')
plt.show()
```

```python
# Residual plot
sns.residplot(x='highway_mpg', y='price', data=df)
plt.title('Residual Plot: Price vs Highway MPG')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np

# Fit a model
X = df[['highway_mpg']]
y = df['price']
model = LinearRegression()
model.fit(X, y)
y_pred = model.predict(X)

# Plot predicted vs actual
sns.kdeplot(y, label='Actual Values', color='red')
sns.kdeplot(y_pred, label='Predicted Values', color='blue')
plt.title('Distribution Plot: Actual vs Predicted Price')
plt.legend()
plt.show()
```
