# 3rd order Polynomial Regression

```python
import numpy as np
import matplotlib.pyplot as plt

# Sample data
x = np.array([1, 2, 3, 4, 5])
y = np.array([100, 150, 300, 600, 1300])  # Non-linear trend

# Fit 3rd order polynomial
coeffs = np.polyfit(x, y, 3)
model = np.poly1d(coeffs)

# Predict
x_new = np.linspace(1, 5, 100)
y_pred = model(x_new)

# Plot
plt.scatter(x, y, label='Actual')
plt.plot(x_new, y_pred, color='red', label='3rd Order Polynomial Fit')
plt.legend()
plt.title('Polynomial Regression (3rd Order)')
plt.show()

# Print model
print("Polynomial equation:")
print(model)
```

```python
import numpy as np
p = np.poly1d([2, 3, 5])  # coefficients in descending order of x
         # represents 2*x**2 + 3*x + 5
print(p)
```

# Multivariate Polynomial Regression

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
import pandas as pd

# Sample 2D dataset
x = pd.DataFrame({'x1': [1, 2, 3], 'x2': [4, 5, 6]})
y = np.array([15, 25, 40])
```

```python
# Create polynomial features (degree=2)
poly = PolynomialFeatures(degree=2, include_bias=False)
x_poly = poly.fit_transform(x)
        # contains [x1, x2, x1^2, x1*x2, x2^2]
print(x_poly)
```

```python
# Fit model
model = LinearRegression()
model.fit(x_poly, y)
```

```python
# Print coefficients
print("Coefficients:", model.coef_)
print("Intercept:", model.intercept_)
```
